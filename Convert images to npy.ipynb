```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
import time

save_dir = '/home/kasatkin/ros_workspaces/esgm_ws/depth_maps_wo_post_npy/'
depth_folder_wo_post = '/home/kasatkin/ros_workspaces/esgm_ws/depth_maps_wo_post/'
depth_folder = '/home/kasatkin/ros_workspaces/esgm_ws/depth_maps/'
```

```python
def transform(arr):
    image = arr.copy()
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i, j] == 0.:
                image[i, j] = -1.
                
    return image


def plot_dm(image):
    fig, ax = plt.subplots()

    ax.imshow(image)

    fig.set_figwidth(15)    #  ширина и
    fig.set_figheight(15)    #  высота "Figure"

    plt.show()
```

```python
img_1 = np.array(Image.open(os.path.join(depth_folder_wo_post, sorted(os.listdir(depth_folder_wo_post))[0])), dtype=np.float32)
plot_dm(img_1)
```

```python
img_2 = np.array(Image.open(os.path.join(depth_folder, sorted(os.listdir(depth_folder))[0])), dtype=np.float32)
plot_dm(img_2)
```
